# Forensic Analysis: Operational Revenue Leakage
**Objective:** Reconcile Service Logs (CRM) with Invoicing Data (ERP) to find unbilled revenue and process bottlenecks.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load reconciled dataset
df = pd.read_csv('data/crm_service_logs.csv')
erp = pd.read_csv('data/erp_invoices.csv')
merged = pd.merge(df, erp, on='Service_ID', how='left')

print(f"Reconciliation Rate: {merged['Invoice_ID'].notnull().mean():.1%}")
```

### Visualizing Billing Lag by Region
We calculate the delta between service completion and invoice creation to identify regional process inefficiency.

```python
merged['Completion_Timestamp'] = pd.to_datetime(merged['Completion_Timestamp'])
merged['Invoice_Created_Timestamp'] = pd.to_datetime(merged['Invoice_Created_Timestamp'])
merged['Lag_Days'] = (merged['Invoice_Created_Timestamp'] - merged['Completion_Timestamp']).dt.days

plt.figure(figsize=(10,6))
sns.boxplot(data=merged, x='Regional_Office', y='Lag_Days', palette='viridis')
plt.title('Regional Billing Lag (Days) - Forensic Identification of Bottlenecks')
plt.ylabel('Days from Completion to Invoice')
plt.show()
```
