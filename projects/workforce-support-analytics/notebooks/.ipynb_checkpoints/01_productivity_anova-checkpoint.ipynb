```python
# 01_productivity_anova.ipynb
# Project: Workforce Support Intervention → Productivity & ROI
# Business Question: Does structured support measurably improve weekly output vs light/no support?

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Data generation (or load from CSV)
np.random.seed(42)
n = 180
support_type = np.repeat(["None", "Light", "Structured"], n//3)
tenure = np.random.normal(3, 1.2, n)
workload = np.random.normal(40, 5, n)
base_output = np.random.normal(70, 10, n)
effect = np.where(support_type == "Structured", 12, np.where(support_type == "Light", 5, 0))
weekly_output = base_output + effect - (workload - 40)*0.5 + tenure*1.2
error_rate = np.random.normal(5, 1.5, n) - effect*0.1
focus_score = np.random.normal(6, 1.2, n) + effect*0.2

df = pd.DataFrame({
    "SupportType": support_type,
    "Tenure": tenure,
    "Workload": workload,
    "WeeklyOutput": weekly_output,
    "ErrorRate": error_rate,
    "FocusScore": focus_score
})

# Quick EDA
print(df.groupby('SupportType')['WeeklyOutput'].agg(['mean', 'std', 'count']))
print("\nDataset shape:", df.shape)
```

```python
# ANOVA: Test if WeeklyOutput differs across support types
model = ols('WeeklyOutput ~ C(SupportType)', data=df).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
print("ANOVA Results:")
print(anova_results)
```

```python
# Tukey post-hoc: Which groups specifically differ?
tukey = pairwise_tukeyhsd(endog=df['WeeklyOutput'], groups=df['SupportType'], alpha=0.05)
print(tukey)
```

```python
# Regression: Does support effect hold controlling for tenure/workload?
reg_model = ols('WeeklyOutput ~ Tenure + Workload + C(SupportType)', data=df).fit()
print(reg_model.summary())
```

```python
# Executive visualization
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="SupportType", y="WeeklyOutput")
plt.title("Weekly Output by Support Type", fontsize=16, fontweight='bold')
plt.ylabel("Weekly Output KPI")
plt.xlabel("Support Model")
plt.tight_layout()
plt.savefig('../visuals/productivity_by_support.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Save data for downstream use
df.to_csv('../data/workforce_productivity.csv', index=False)
print("Data saved to workforce_productivity.csv")
```
