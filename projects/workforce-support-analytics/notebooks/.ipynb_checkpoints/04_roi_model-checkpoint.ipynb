```python
# 04_roi_model.ipynb
# Business Question: What is the financial impact of support programs?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from previous notebooks
df = pd.read_csv('../data/workforce_productivity.csv')
df3 = pd.read_csv('../data/program_impact_timeseries.csv')

# Productivity lift by support type
group_means = df.groupby("SupportType")["WeeklyOutput"].mean()
print("Weekly Output by Support Type:")
print(group_means)

# ROI calculation
productivity_lift = group_means["Structured"] - group_means["None"]
value_per_unit = 150  # Assumed business value per KPI unit
weekly_value_lift = productivity_lift * value_per_unit
annual_value_lift = weekly_value_lift * 52
program_cost = 3000  # Annual cost per employee

roi = (annual_value_lift - program_cost) / program_cost
print(f"\nStructured vs None Lift: {productivity_lift:.1f} units/week")
print(f"Annual Value per Employee: ${annual_value_lift:,.0f}")
print(f"ROI: {roi:.0%}")
```

```python
df.to_csv('../data/roi_model.csv', index=False)
print("Data saved to roi_model.csv")
```
